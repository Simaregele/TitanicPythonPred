Поделим данные на трэйн тест

```python
df_targ = df['Survived']
df_data = df.drop(columns=["Survived"])
```

```python
df_data = df_data.select_dtypes(exclude=['object'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data, df_targ, test_size=0.33, random_state=42)
```

regression lets try it

```python
y_test
```

```python
logreg = LogisticRegression(solver='liblinear', random_state=0)
```

```python
X_train
```

```python
logreg.fit(X_train, y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
metrics.precision_score(y_test, y_pred)
```

```python
logreg.get_paramst_params()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)
err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)
print(err_train, err_test)
```
